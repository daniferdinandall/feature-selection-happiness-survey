```python
# Import modules
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('../datasets/dataset.csv')
df.head()
```

```python
# Selecting the columns that we want to use

df_selected = df[['B4K4', 'B4K5', 'B4K6', 'R501', 'R502', 'R503', 'R601A', 'R603', 'R604A', 'R604B', 'R604C', 'R604D', 'R604E', 'R604F', 'R604G', 'R609A', 'R609B', 'R701A', 'R702A', 'R705A', 'R801A', 'R801B', 'R801C', 'R801D', 'R901', 'R902A', 'R1101A1', 'R1101A2', 'R1101A3', 'R1101A4', 'R1101A5', 'R1101B1', 'R1101B2', 'R1101B3', 'R1101C', 'R1201', 'R1203', 'R1204', 'R1205', 'R1206A', 'R1206B', 'R1207', 'R1208', 'R1209', 'R1210A', 'R1210B', 'R1210C', 'R1210D', 'R1210E', 'R1005', 'R1302', 'R1304', 'R1306', 'R505', 'R601B2', 'R605', 'R610', 'R708', 'R1211', 'R802', 'R903', 'R1102', 'R1108', 'R1402', 'R1404', 'R1406', 'R1408', 'R1410', 'R1412', 'WEIGHT']].copy()

df_selected.shape
```

```python
# membuat kolom baru dari kolom R605 berdasarkan R601A = 1 atau R605 berdasarkan R601B2 = 2
df_selected['IPUKU'] = np.where(df_selected['R601A'] == 1, df_selected['R605'], df_selected['R601B2'])

# drop kolom R605 dan R601B2
df_selected = df_selected.drop(['R605', 'R601B2'], axis=1)
df_selected[['R601A', 'IPUKU']].head()
```

Index Kepuasan Hidup Personal

index_khp = (R505 * 18.34 + IPUKU * 21.67 + R610 * 22.81 + R708 * 17.04 + R1211 * 20.14)

Index Kepuasan Hidup Sosial

index_khs = (R802 * 19.41 + R903 * 18.93 + R1005 * 22.13 + R1102 * 20.64 + R1108 * 18.89)

Index Kepuasan Hidup

index_kh = (index_khp * 0.5 + index_khs * 0.5)

index_p = (R1302 * 25.86 + R1304 * 36.80 + R1306 * 37.34)

Index Makna Hidup

Index Kebahagiaan

index_k = (index_kh * 0.348 + index_p * 0.3118 + index_mh * 0.3402)


```python
# Menghitung Index Kepuasan Hidup Personal
df_selected['index_khp'] = round((
    df_selected['R505'] * 0.1834 +
    df_selected['IPUKU'] * 0.2167 +
    df_selected['R610'] * 0.2281 +
    df_selected['R708'] * 0.1704 +
    df_selected['R1211'] * 0.2014
) * 10 , 2)

# Menghitung Index Kepuasan Hidup Sosial
df_selected['index_khs'] = round((
    df_selected['R802'] * 0.1941 +
    df_selected['R903'] * 0.1893 +
    df_selected['R1005'] * 0.2213 +
    df_selected['R1102'] * 0.2064 +
    df_selected['R1108'] * 0.1889
) * 10 , 2)

# Menghitung Index Kepuasan Hidup
df_selected['index_kh'] = round((
    df_selected['index_khp'] * 0.5 +
    df_selected['index_khs'] * 0.5
), 2)

# Menghitung Index Perasaan
df_selected['index_p'] = round((
    df_selected['R1302'] * 0.2586 +
    df_selected['R1304'] * 0.3680 +
    df_selected['R1306'] * 0.3734
) * 10 , 2)

# Menghitung Index Makna Hidup
df_selected['index_mh'] = round((
    df_selected['R1402'] * 0.1656 +
    df_selected['R1404'] * 0.1844 +
    df_selected['R1406'] * 0.1527 +
    df_selected['R1408'] * 0.1548 +
    df_selected['R1410'] * 0.1748 +
    df_selected['R1412'] * 0.1678
) * 10 , 2)

# Menghitung Index Kebahagiaan
df_selected['index_k'] = round((
    df_selected['index_kh'] * 0.348 +
    df_selected['index_p'] * 0.3118 +
    df_selected['index_mh'] * 0.3402
), 2)
```

```python
# Menyimpan hasil perhitungan index kebahagiaan ke dalam file csv
df_selected.to_csv('../datasets/index_kebahagiaan.csv', index=False)
```
