```python
# Import modules
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('../datasets/dataset.csv')
df.head()
```

```python
# drop unnecessary columns
df.drop(['URUT','R101','R101_NAMA','R105','R108','R111','R201','R201A','R201B','B4K1','B4K3','B4K4','B4K5','B4K6','B4K7','LAMA_TINGG','R1501','R1502'], axis=1, inplace=True)
df.shape
```

```python
# membuat kolom baru dari kolom R605 berdasarkan R601A = 1 atau R605 berdasarkan R601B2 = 2
df['IPUKU'] = np.where(df['R601A'] == 1, df['R605'], df['R601B2'])

# drop kolom R605 dan R601B2
# df = df.drop(['R605', 'R601B2'], axis=1)
df[['R601A', 'IPUKU']].head()
```

index_khp = (R505 * 18.34 + IPUKU * 21.67 + R610 * 22.81 + R708 * 17.04 + R1211 * 20.14)

index_khs = (R802 * 19.41 + R903 * 18.93 + R1005 * 22.13 + R1102 * 20.64 + R1108 * 18.89)

index_p = (R1302 * 25.86 + R1304 * 36.80 + R1306 * 37.34)

index_k = (index_kh * 0.348 + index_p * 0.3118 + index_mh * 0.3402)


```python
# Menghitung Index Kepuasan Hidup Personal
df['index_khp'] = round((
    df['R505'] * 0.1834 +
    df['IPUKU'] * 0.2167 +
    df['R610'] * 0.2281 +
    df['R708'] * 0.1704 +
    df['R1211'] * 0.2014
) * 10 , 2)

# Menghitung Index Kepuasan Hidup Sosial
df['index_khs'] = round((
    df['R802'] * 0.1941 +
    df['R903'] * 0.1893 +
    df['R1005'] * 0.2213 +
    df['R1102'] * 0.2064 +
    df['R1108'] * 0.1889
) * 10 , 2)

# Menghitung Index Kepuasan Hidup
df['index_kh'] = round((
    df['index_khp'] * 0.5 +
    df['index_khs'] * 0.5
), 2)

# Menghitung Index Perasaan
df['index_p'] = round((
    df['R1302'] * 0.2586 +
    df['R1304'] * 0.3680 +
    df['R1306'] * 0.3734
) * 10 , 2)

# Menghitung Index Makna Hidup
df['index_mh'] = round((
    df['R1402'] * 0.1656 +
    df['R1404'] * 0.1844 +
    df['R1406'] * 0.1527 +
    df['R1408'] * 0.1548 +
    df['R1410'] * 0.1748 +
    df['R1412'] * 0.1678
) * 10 , 2)

# Menghitung Index Kebahagiaan
df['index_k'] = round((
    df['index_kh'] * 0.348 +
    df['index_p'] * 0.3118 +
    df['index_mh'] * 0.3402
), 2)
```
